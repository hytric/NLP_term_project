```python
from pathlib import Path

import pandas as pd
from IPython.display import display
from transformers import XLMRobertaTokenizer

TOKENIZER_DIR = Path("/home/axt/mnt2/jongha/second_try/artifacts/13_v2_tokenizer/tokenizers/xlmr_v2_target10_added_32000")
tokenizer = XLMRobertaTokenizer.from_pretrained(str(TOKENIZER_DIR), local_files_only=True)

BASE_VOCAB_SIZE = tokenizer.vocab_size
print(f"unk_token={tokenizer.unk_token!r}, unk_token_id={tokenizer.unk_token_id}")
print(f"tokenizer_dir={TOKENIZER_DIR}")
print(f"base_vocab_size={BASE_VOCAB_SIZE}, len_tokenizer={len(tokenizer)}, added_tokens={len(tokenizer) - BASE_VOCAB_SIZE}")
```

```python
TEXT = """ⲘⲠⲈⲢϮϨⲀⲠ ϨⲒⲚⲀ ⲚⲦⲞⲨϢⲦⲈⲘϮϨⲀⲠ ⲈⲢⲰⲦⲈⲚ."""

tokens = tokenizer.tokenize(TEXT)
ids = tokenizer.convert_tokens_to_ids(tokens)

table = pd.DataFrame([
    {
        "idx": idx,
        "visible_token": token,
        "token_id": int(token_id),
        "id_maps_back_to": tokenizer.convert_ids_to_tokens(int(token_id)),
        "is_visible_unk": token == tokenizer.unk_token,
        "is_id_level_unk": token_id == tokenizer.unk_token_id,
        "is_base_vocab_id": token_id < BASE_VOCAB_SIZE,
        "is_added_token_id": token_id >= BASE_VOCAB_SIZE,
    }
    for idx, (token, token_id) in enumerate(zip(tokens, ids), start=1)
])

summary = pd.DataFrame([{
    "text": TEXT,
    "token_count": len(tokens),
    "visible_unk_count": int(table["is_visible_unk"].sum()) if len(table) else 0,
    "id_level_unk_count": int(table["is_id_level_unk"].sum()) if len(table) else 0,
}])

display(summary)
display(table)
```
